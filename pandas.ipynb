```python
 1. Reading and Saving Data
python
Copy
Edit
import pandas as pd

df = pd.read_csv('file.csv')         # Load data from a CSV
df.to_csv('new_file.csv')            # Save DataFrame to CSV
2. Basic Information
python
Copy
Edit
df.head()        # First 5 rows
df.tail()        # Last 5 rows
df.info()        # Overview (columns, data types, non-null values)
df.shape         # (rows, columns)
df.columns       # List of column names
df.dtypes        # Data types of each column
3. Selecting Data
python
Copy
Edit
df['Age']                   # Select single column
df[['Name', 'Age']]         # Select multiple columns
df.loc[0]                   # Select row by index label
df.iloc[0]                  # Select row by position
4. Filtering Rows (Conditionals)
python
Copy
Edit
df[df['Age'] > 30]           # Filter rows where Age > 30
df[df['City'] == 'Delhi']    # Filter rows by category
5. Sorting Data
python
Copy
Edit
df.sort_values('Age')                    # Sort by Age
df.sort_values(['Age', 'Salary'])        # Sort by multiple columns
 6. Handling Missing Data
python
Copy
Edit
df.dropna()          # Remove rows with any missing values
df.fillna(0)         # Fill missing values with 0
df.isnull().sum()    # Count missing values per column
7. Aggregating and Grouping
python
Copy
Edit
df['Salary'].mean()                           # Average salary
df.groupby('City').mean()                     # Mean by city
df.groupby('Department')['Salary'].sum()      # Sum salary per dept
 8. Adding and Removing Columns
python
Copy
Edit
df['NewCol'] = df['Salary'] * 2       # Add new column
df = df.drop('NewCol', axis=1)        # Delete column
 9. Merging and Joining Data
python
Copy
Edit
df1 = pd.read_csv('file1.csv')
df2 = pd.read_csv('file2.csv')
pd.concat([df1, df2])                     # Stack DataFrames

               # Join by column
 10. Exporting Final Data
df.to_csv('clean_data.csv', index=False)  # Save without index
```
